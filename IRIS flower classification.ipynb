This is a simple **LOGISTIC REGRESSION** MODEL to classify Iris Species


Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

Loading **CSV file**

```python
df = pd.read_csv("Iris.csv")
df.head()
```

```python
#Preparing Training set
#X =feature values all columns accept last
x =df.iloc[:,:-1]
# Y stores target values which is the last column (Species)
y =df.iloc[:,-1]

```

**Plot** the **target values** against **feature values**

```python
plt.xlabel('Features')
plt.ylabel('Species')

pltX = df.loc[:,'SepalLengthCm']
pltY = df.loc[:,'Species']
plt.scatter(pltX,pltY,color ='green', label='SepalLengthCm')

pltX =df.loc[:,'SepalWidthCm']
pltY =df.loc[:,'Species']
plt.scatter(pltX,pltY,color ='red',label ='SepalWidthCm')

pltX = df.loc[:,'PetalLengthCm']
pltY = df.loc[:,'Species']
plt.scatter(pltX,pltY,color ='blue',label ='PetalLengthCm')

pltX =df.loc[:,'PetalWidthCm']
pltY =df.loc[:,'Species']
plt.scatter(pltX,pltY,color ='yellow',label ='PetalWidthCm')

plt.legend()
plt.show()










```

Split data in 80% TRAINING and 20% TESTING

```python
from pandas.core.common import random_state
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2,random_state =42)
```

```python
#Train model
model= LogisticRegression()
model.fit(x_train,y_train)
```

```python
#Test model
predictions = model.predict(x_test)
print(predictions)
print(y_test)
```

Precission,Recall, f1-score

```python
print(classification_report(y_test,predictions))
```

```python
#Accuracy
print(accuracy_score(y_test,predictions))
```
